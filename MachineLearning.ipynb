```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv('src/datasets/transacaoCartao.csv', sep=';', decimal=',')
df['datatrimestre'] = pd.to_datetime(df['datatrimestre'])
```

Correção de ponto e vírgula nos campos de valor:

```python
for col in df.columns:
    if col.startswith('valor'):
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        df[col] = df[col].astype(float)
```

Para garatir que os dados estão ordenados:

```python
df = df.sort_values(by='datatrimestre')
```

### Análise de Correlação

A correlação é uma medida estatística que indica a força e a direção da relação entre duas variáveis. Quando duas variáveis estão correlacionadas, significa que mudanças em uma tendem a estar associadas a mudanças na outra.

- Se a correlação é positiva, as duas variáveis aumentam ou diminuem juntas.

- Se é negativa, quando uma sobe, a outra tende a descer.

- Uma correlação próxima de zero indica pouca ou nenhuma relação linear entre elas.

A forma mais comum de medir correlação é o coeficiente de correlação de Pearson, que varia de -1 a 1:

- 1 = correlação perfeitamente positiva

- 0 = nenhuma correlação linear

- -1 = correlação perfeitamente negativa

**Importante**: correlação não implica causalidade — duas variáveis podem estar relacionadas sem que uma cause a outra.

```python
correlacao = df[[col for col in df.columns if col.startswith('valor')]].corr()
sns.heatmap(correlacao)
```

Como estamos tentando prever o gasto com Cartão de Crédito, podemos limitar o estudo de correção somente a variável de interesse:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao[['valorCartaoCredito']].sort_values(by='valorCartaoCredito', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlação dos Valores com Cartão de Crédito')
plt.tight_layout()
plt.show()
```

### Random Forest

O Random Forest é um algoritmo de aprendizado de máquina baseado em conjuntos de árvores de decisão. Ele é utilizado para tarefas de classificação e regressão, combinando múltiplas árvores para gerar previsões mais precisas e estáveis.

- O modelo cria várias árvores de decisão independentes, cada uma treinada com uma amostra aleatória dos dados (técnica chamada *bagging*).

- A cada divisão em uma árvore, apenas um subconjunto aleatório de variáveis é considerado, o que aumenta a diversidade entre as árvores.

- A previsão final é feita por votação (para classificação) ou média (para regressão) dos resultados das árvores.

As principais características do Random Forest incluem:

- **Robustez contra overfitting**, pois a combinação de múltiplas árvores reduz o risco de ajuste excessivo aos dados de treino.

- **Capacidade de lidar com dados de alta dimensionalidade**, variáveis categóricas e dados ausentes.

**Importante**: o Random Forest é um modelo poderoso, mas menos interpretável do que uma única árvore de decisão. Além disso, seu desempenho pode depender da escolha de parâmetros como o número de árvores e a profundidade máxima permitida.

```python
X_import = df.drop(columns=['valorCartaoCredito', 'datatrimestre'])
Y_import = df['valorCartaoCredito']

modelo_temp = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_temp.fit(X_import, Y_import)

importancias = pd.Series(modelo_temp.feature_importances_, index=X_import.columns)
importantes = importancias.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 8))
importantes.plot(kind='bar')
plt.title('Importância das Variáveis (Random Forest)')
plt.tight_layout()
plt.show()
```

```python
melhores_variaveis = importantes.head(8).index.tolist()

df['lag1'] = df['valorCartaoCredito'].shift(1)
df['lag2'] = df['valorCartaoCredito'].shift(2)

df_model = df.dropna()

X = df_model[melhores_variaveis + ['lag1', 'lag2']]
Y = df_model['valorCartaoCredito']

X_train, X_test, y_train, y_test = train_test_split(X, Y, shuffle=False, test_size=0.3)

modelo = RandomForestRegressor(
    n_estimators=100, 
    random_state=42, 
    max_depth=4, 
    min_samples_split=3, 
    min_samples_leaf=2,
    max_features='log2'
)
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse:.2f}')

y_full = np.concatenate([modelo.predict(X_train), y_pred])
datas_full = df_model['datatrimestre'].iloc[-len(y_full):]
plt.figure(figsize=(12, 6))
plt.plot(df_model['datatrimestre'], df_model['valorCartaoCredito'], label='Valor Real (Total)', color='blue')
plt.plot(datas_full, y_full, label='Valor Previsto (Modelo)', color='green', linestyle='--')
plt.title('Previsão do valorCartaoCredito - Toda a Série')
plt.xlabel('Data')
plt.ylabel('Valor (R$)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Esse foi o primeiro modelo obtido, perceba que ele se adequa bem aos valores reais até certo momento, depois começa a errar.

### Previsão com Prophet

O **Prophet** é uma biblioteca desenvolvida pela Meta (Facebook) para realizar previsões de séries temporais de forma precisa e intuitiva. É especialmente eficaz para dados com padrões sazonais e tendências não lineares ao longo do tempo.

O Prophet utiliza um modelo aditivo onde os componentes principais são somados:

> y(t) = tendência(t) + sazonalidade(t) + feriados(t) + erro

- **Tendência**: crescimento linear ou logístico ao longo do tempo.

- **Sazonalidade**: variações que se repetem ao longo de ciclos diários, semanais ou anuais.

- **Feriados/eventos especiais**: permite adicionar datas importantes que podem impactar a série.

A biblioteca é projetada para ser:

- **Fácil de usar**: exige poucos ajustes e aceita dados em um formato simples (colunas `ds` para data e `y` para valor).

- **Interativa**: permite ajustes manuais nos componentes sazonais, tendências e feriados.

- **Robusta**: lida bem com dados ausentes, grandes intervalos de tempo e mudanças de tendência.

**Importante**: embora o Prophet seja poderoso para séries temporais com padrões claros, ele assume que esses padrões continuarão no futuro. Interpretação cuidadosa e validação são fundamentais para boas previsões.

```python
from prophet import Prophet
import pandas as pd


df_prophet = df[['datatrimestre', 'valorCartaoCredito']].rename(columns={
    'datatrimestre': 'ds',
    'valorCartaoCredito': 'y'
})


model = Prophet()
model.fit(df_prophet)

future = model.make_future_dataframe(periods=4, freq='QE')
forecast = model.predict(future)
fig = model.plot(forecast)
```

```python
df_prophet = df[['datatrimestre', 'valorCartaoCredito']].rename(columns={
    'datatrimestre': 'ds',
    'valorCartaoCredito': 'y'
})

# Define limite superior ("cap") e inferior ("floor") para a série
max_val = df_prophet['y'].max()
df_prophet['cap'] = max_val * 1.2
df_prophet['floor'] = 0

model = Prophet(
    growth='logistic',
    changepoint_prior_scale=0.3,     # Mais flexível para mudanças de tendência
    seasonality_prior_scale=10.0     # Mais liberdade para curva sazonal
)

model.add_seasonality(name='trimestral', period=365.25/4, fourier_order=5)


model.fit(df_prophet)

future = model.make_future_dataframe(periods=4, freq='QE')
future['cap'] = df_prophet['cap'].max()
future['floor'] = 0

forecast = model.predict(future)

fig1 = model.plot(forecast)
plt.title("Previsão valorCartaoCredito com tendência logística e sazonalidade trimestral")
plt.xlabel("Data")
plt.ylabel("Valor (R$)")
plt.tight_layout()
plt.show()


fig2 = model.plot_components(forecast)
plt.tight_layout()
plt.show()
```

Com o modelo ajustado, podemos prever valores futuros:

```python
future_only = forecast[forecast['ds'] > df_prophet['ds'].max()]
print(future_only[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

model = Prophet(
    growth='logistic',
    changepoint_prior_scale=0.3,
    seasonality_prior_scale=10.0
)
model.add_seasonality(name='trimestral', period=365.25/4, fourier_order=5)
model.fit(df_prophet)


future = model.make_future_dataframe(periods=4, freq='QE')  
future['cap'] = df_prophet['cap'].max()
future['floor'] = 0
```

```python
forecast = model.predict(future)


ultima_data_real = df_prophet['ds'].max()
historico = df_prophet.copy()
forecast_modelado = forecast[forecast['ds'] <= ultima_data_real]
forecast_futuro = forecast[forecast['ds'] > ultima_data_real]

ultimo_modelado = forecast_modelado.iloc[-1][['ds', 'yhat']]
primeiro_previsto = forecast_futuro.iloc[0][['ds', 'yhat']]

transicao = pd.DataFrame({
    'ds': [ultimo_modelado['ds'], primeiro_previsto['ds']],
    'yhat': [ultimo_modelado['yhat'], primeiro_previsto['yhat']]
})
```

```python
plt.figure(figsize=(12, 6))

plt.plot(historico['ds'], historico['y'], label='Real (Histórico)', color='blue', marker='o')
plt.plot(forecast_modelado['ds'], forecast_modelado['yhat'], label='Modelado', color='green', linestyle='--')
plt.plot(forecast_futuro['ds'], forecast_futuro['yhat'], label='Previsão', color='red', linestyle='--', marker='o')

plt.fill_between(forecast_futuro['ds'], forecast_futuro['yhat_lower'], forecast_futuro['yhat_upper'],
                 color='red', alpha=0.2, label='Intervalo de Confiança')


plt.plot(transicao['ds'], transicao['yhat'], color='red', linestyle='--')  # linha de conexão

plt.title('Valor do Cartão de Crédito - Real, Modelado e Previsão (Prophet)')
plt.xlabel('Data')
plt.ylabel('Valor (R$)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
