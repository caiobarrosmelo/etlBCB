```python
(df_cards['valorCartaoPrePago']/df_cards['quantidadeCartaoPrePago']).mean()
```

 A diferença existe porque, na segunda abordagem de cada série, está dando peso igual a cada trimestre, independentemente de quantas transações ocorreram naquele período. Já na primeira, cada transação — seja no trimestre que for — “conta” igualmente, o que faz o trimestre com mais transações influenciar mais o resultado final.

 | Métrica                                  | Fórmula                                                    | Interpretação                                                  |
| ---------------------------------------- | ---------------------------------------------------------- | -------------------------------------------------------------- |
| Média geral (ponderada)                  | $\sum_i valorPix_i \,/\, \sum_i quantidadePix_i$           | Valor médio por transação considerando todo o volume           |
| Média das médias trimestrais (não-pond.) | $\tfrac{1}{n}\sum_i\bigl(valorPix_i/quantidadePix_i\bigr)$ | Valor médio por transação onde cada trimestre tem “peso” igual |

Use a **média geral** quando quiser o valor médio de toda a amostra, e a **média das médias trimestrais** quando quiser tratar cada período (trimestre) com igual importância.

Box-Plot comparativo do Valor Médio por Transação dos Cartões de Crédito, Débito e Pré-Pago

```python
import matplotlib.pyplot as plt

plt.figure()

# Ajustar as posições para os boxplots
plt.boxplot([x1.dropna(), x2.dropna(), x3.dropna()], positions=[1, 2, 3])

# Melhorar a aparência
plt.xticks([1, 2, 3], ['Cartão de Crédito', 'Cartão de Débito', 'Cartão Pré-Pago'], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Adicionar título e rótulos
plt.title('Box-plot do valor médio por transação dos Cartões de Crédito, Débito e Pré-Pago', fontsize=14)
plt.xlabel('Cartões', fontsize=12)
plt.ylabel('Valor médio por transação (R$)', fontsize=12)

# Layout ajustado
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

**Boxplot do valor médio por transação por tipo de cartão**

O gráfico apresenta a distribuição dos valores médios por transação, separados por tipo de cartão: Crédito, Débito e Pré-Pago.

**Cartão de Crédito**  
Esse tipo de cartão apresenta os maiores valores médios por transação. A mediana está acima de R$ 125, indicando que a metade das transações ocorre com valores superiores a esse. A variação entre os quartis é pequena, mostrando uma distribuição relativamente estável. Poucos valores extremos foram observados, o que sugere consistência no uso do cartão de crédito para compras de valor mais elevado.

**Cartão de Débito**  
A mediana das transações com débito está por volta de R$ 62. A distribuição é mais compacta e apresenta menor dispersão em comparação com o crédito. Há alguns outliers positivos, indicando que, em casos pontuais, há transações com valores maiores. O débito parece ser utilizado principalmente para compras de valor intermediário e com maior controle.

**Cartão Pré-Pago**  
É o cartão com os menores valores médios por transação. A mediana está em torno de R$ 27, mas há vários outliers acima de R$ 50, inclusive alguns próximos a R$ 85. Isso indica que, embora normalmente utilizado para valores baixos, esse tipo de cartão também é usado eventualmente para valores mais altos, possivelmente quando há concentração de saldo.

Em resumo, observa-se um padrão claro: o cartão de crédito é usado para transações de maior valor, o de débito para valores intermediários e o pré-pago para valores menores, com algumas exceções.
